```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from utils import LogisticRegression, accuracy_score
```

```python
data = pd.read_csv("cancer.csv")
```

```python
data.dtypes
```

```python
# X_data = data.drop(data.loc[:, "concave points_mean":"fractal_dimension_worst"].columns, axis=1)
X = data.drop(columns=["diagnosis(1=m, 0=b)"])
X
```

```python
y = data["diagnosis(1=m, 0=b)"]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
log_reg = LogisticRegression()
log_reg.train_data(X_train, y_train)
```

```python
y_prediction = log_reg.predict(X_test)
```

```python
res = accuracy_score(y_test, y_prediction)
res
```
