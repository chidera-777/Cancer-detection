```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Normalization, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR)
tf.autograph.set_verbosity(0)
# from utils import LogisticRegression, accuracy_score
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/cancer.csv")
```

```python
data.isnull().sum()
```

```python
# X_data = data.drop(data.loc[:, "concave points_mean":"fractal_dimension_worst"].columns, axis=1)
X = data.drop(columns=["diagnosis(1=m, 0=b)"])
X
```

```python
y = data["diagnosis(1=m, 0=b)"]
y
```

```python
X_norm = Normalization(axis=-1)
X_norm.adapt(X)
Xn = X_norm(X)
Xn = pd.DataFrame(Xn)
Xn
```

```python
seed = 20
np.random.seed(seed)
X_train, X_, y_train, y_ = train_test_split(Xn, y, test_size=0.40, random_state=seed)
X_val, X_test, y_val, y_test = train_test_split(X_, y_, test_size=0.50)
print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
print(y_train.shape)
print(y_val.shape)
print(y_test.shape)
print(X_train.head())
```

# **Let's test with these three models to see which one will be perfect to train the model**

```python
tf.random.set_seed(20)
model_1 = Sequential(
    [
        Dense(40, activation="relu"),
        Dense(25, activation="relu"),
        Dense(30, activation="relu"),
        Dense(10, activation="relu"),
        Dense(1, activation="linear")
    ], name = "model_1"
)

model_2 = Sequential(
    [
        Dense(33, activation="relu"),
        Dense(25, activation="relu"),
        Dense(10, activation="relu"),
        Dense(20, activation="relu"),
        Dense(15, activation="relu"),
        Dense(10, activation="relu"),
        Dense(1, activation="linear")
    ], name = "model_2"
)

model_3 = Sequential(
    [
        Dense(25, activation="relu"),
        Dense(10, activation="relu"),
        Dense(1, activation="linear")
    ], name = "model_3"
)

model_list = [model_1, model_2, model_3]
```

```python
def accuracy_score(y, y_pred):
    res = np.mean(y_pred == y) * 100
    return res
```

```python
train_error = []
dev_error = []
train_accuracy = []
dev_accuracy = []

for model in model_list:
  model.compile(
      loss = BinaryCrossentropy(from_logits=True),
      optimizer = Adam(learning_rate=0.001)
  )

  print(f"Training {model.name}....")


  model.fit(
      X_train, y_train,
      epochs=300,
      verbose=0
  )

  print("Done!!!")

  y_pred = model.predict(X_train)
  y_pred = tf.math.sigmoid(y_pred)
  yhat = np.where(y_pred >= 0.5, 1, 0)
  yhat = yhat.reshape(-1)
  print(f"train errors = {np.count_nonzero(yhat != y_train)}")
  train_err = np.mean(yhat != y_train)
  train_error.append(train_err)
  train_accuracy.append(accuracy_score(y_train, yhat))


  y_pred = model.predict(X_val)
  y_pred = tf.math.sigmoid(y_pred)
  yhat = np.where(y_pred >= 0.5, 1, 0)
  yhat = yhat.reshape(-1)
  print(f"val errors = {np.count_nonzero(yhat != y_val)}")
  dev_err = np.mean(yhat != y_val)
  dev_error.append(dev_err)
  dev_accuracy.append(accuracy_score(y_val, yhat))


for num in range(len(train_error)):
  print(
      f"Training error for model {num+1}: {train_error[num]:.5f} .......  "
      f"Validation error for model {num+1}: {dev_error[num]:.5f} .......  "
  )
```

**From this, we can see that model_2 does better than the other 2, although they all do well, model_2 out-performs them**

**Therefore we are going to be using model_2 as our training model**

```python
y_pred = model_list[1].predict(X_test)
y_pred = tf.math.sigmoid(y_pred)
yhat = np.where(y_pred >= 0.5, 1, 0)
yhat = yhat.reshape(-1)
print(f"test errors = {np.count_nonzero(yhat != y_test)}")
test_err = np.mean(yhat != y_test)
test_accuracy = accuracy_score(y_test, yhat)


print(f"Selected model: {2}")
print(f"Training error: {train_error[1]:.5f}...... Training accuracy: {train_accuracy[1]:.3f}%")
print(f"Validation error: {dev_error[1]:.5f}...... Validation accuracy: {dev_accuracy[1]:.3f}%")
print(f"Test error: {test_err:.5f}....... Test accuracy: {test_accuracy:.3f}%")
```
